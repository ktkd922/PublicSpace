# A → B → C で表される逐次反応の解析解を plot

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
a0 = 1
k = 1
t0 = 0
t1 = 20

t = np.linspace(t0, t1, 100)
a = a0 * np.exp(-k*t)
b = a0*k*t*np.exp(-k*t)
c = a0*(1-(k*t+1)*np.exp(-k*t))

plt.plot(t, a, label='A')
plt.plot(t, b, label='B')
plt.plot(t, c, label='C')
plt.legend()
plt.show()
```

k1 != k2
の場合

```python
k1 = 1
k2 = 0.1

a = a0*np.exp(-k1*t)
b = a0*k1/(k2-k1)*(np.exp(-k1*t) - np.exp(-k2*t))
c = a0*(1+(k2*np.exp(-k1*t) - k1*np.exp(-k2*t))/(k1 - k2))

plt.plot(t, a, label='A')
plt.plot(t, b, label='B')
plt.plot(t, c, label='C')
plt.legend()
plt.show()
```
