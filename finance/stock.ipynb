```python
import pandas as pd
from pandas_datareader import data

# stk_dow = data.DataReader('^DJI', 'yahoo', start='2010-01-01', end='2021-12-31')
stk_dow = data.DataReader('^DJI', 'yahoo')
stk_aapl = data.DataReader('AAPL', 'yahoo')
stk_shinetsu = data.DataReader('4063.T', 'yahoo')
stk_shinetsu.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

plt.plot(stk_dow.index, stk_dow['Adj Close']/100, label='DOW/100')
plt.plot(stk_aapl.index, stk_aapl['Adj Close'], label='AAPL')
plt.plot(stk_shinetsu.index, stk_shinetsu['Adj Close']/70, label='ShinEtsu/70')
plt.legend()
plt.show()
```

```python
stk_dow['Adj Close'].corr(stk_aapl['Adj Close'])
```

```python
new_df = pd.concat([stk_dow['Adj Close'], stk_aapl['Adj Close'], stk_shinetsu['Adj Close']], axis=1)
new_df.columns = ['DOW', 'AAPL', 'ShinEtsu']
new_df.tail()
```
