#　高分子の分子量シミュレーション

## 1. 縮重合で得られるポリマーの分子量分布
反応度 $ p $

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn; seaborn.set()
plt.rcParams["font.family"] = "Times"
```

```python
# モノマーを等量反応させた時
p = np.arange(0.9, 0.999, 0.001)
x_n = 1/(1-p)
x_w = (1+p)/(1-p)

plt.tight_layout()
plt.tick_params(labelsize=14)
plt.plot(p, x_n, label="x_n")
plt.plot(p, x_w, label="x_w")
plt.legend()
plt.ylim(0, 500)
plt.xlabel(r'$ p $', fontsize=14)
plt.ylabel(r'$ x_n $', fontsize=14)
plt.show()
```

```python
p_ls = [0.95, 0.96, 0.97, 0.98, 0.98, 0.99]
x = np.arange(0, 500, 5)

# 数平均重合度
for p in p_ls:
    n_x = p**(x-1) * ((1 - p)**2)
    plt.tight_layout()
    plt.tick_params(labelsize=14)
    plt.plot(x, n_x, label="p = " + str(p))

plt.xlabel(r'$ x $', fontsize=14)
plt.ylabel(r'$ n_x $', fontsize=14)
plt.legend()
plt.show()

# 重量平均重合度
for p in p_ls:
    w_x = x*p**(x-1) * ((1 - p)**2)
    plt.tight_layout()
    plt.tick_params(labelsize=14)
    plt.plot(x, w_x, label="p = " + str(p))

plt.xlabel(r'$ x $', fontsize=14)
plt.ylabel(r'$ w_x $', fontsize=14)
plt.legend()
plt.show()
```

## 2. 二種の二官能基モノマーを用いて逐次重合を行う時の数平均重合度

反応率 $ \quad p = 1 $\
仕込み比 $ \quad r = N_A / N_B(< 1) $\
数平均重合度 $ \quad x_n = \frac{1+r}{1-r} $

```python
r = np.arange(0, 1, 0.01)
x_n = (1 + r) / (1 - r)

plt.tight_layout()
plt.tick_params(labelsize=14)
plt.plot(r, x_n)
plt.xlabel(r'$r$', fontsize=14)
plt.ylabel(r'$x_n$', fontsize=14)
plt.ylim(0, 50)
plt.show()
```
